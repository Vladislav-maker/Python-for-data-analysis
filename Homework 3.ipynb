```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
# описание типов данных
df.info()
```

```python
# информация о количестве пропусков
df.isna().sum()
```

```python
# Различные значения по каждому из параметров
df.describe()
```

```python
# Количество пассажиров каждого класса
df['Pclass'].value_counts()
```

```python
# Количество выживших в каждом классе
df.groupby('Pclass')['Survived'].sum()
```

```python
# Процент выживаемости в каждом классе 
df.groupby('Pclass')['Survived'].sum() / df['Pclass'].value_counts() * 100
```

```python
# Извлечение имени 
df['FirstName'] = df['Name'].str.extract(r',\s*[^.]*\.\s*(.*)')
```

```python
# Популярные имена среди мужчин в порядке убывания
df[df['Sex'] == 'male']['FirstName'].value_counts()
```

```python
# Популярные имена среди девушек в порядке убывания
df[df['Sex'] == 'female']['FirstName'].value_counts()
```

```python
# Самые популярные мужские и женские имена в каждом классе
result = (
    df.groupby(['Pclass', 'Sex'])['FirstName'].agg(
        MostCommonName=lambda x: x.value_counts().idxmax(),
        Count=lambda x: x.value_counts().max()
    ).reset_index()
)

print(result)
```
