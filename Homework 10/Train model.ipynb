```python
import pandas as pd

from sklearn.datasets import load_breast_cancer, make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge

# from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

from sklearn.metrics import f1_score, fbeta_score, classification_report
from sklearn.metrics import root_mean_squared_error
```

```python
df = pd.read_csv('realty_data.csv')
df.head()
```

```python
columns = ['product_name', 'period', 'postcode', 'address_name', 'lat', 'lon', 'object_type', 'city', 'settlement', 'district', 'area', 'description', 'source']
df = df.drop(columns = columns)
```

```python
df.head()
```

```python
df['rooms'] = df['rooms'].fillna(value=1)
df.head()
```

```python
df = df.dropna()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns=['price', 'floor']),
    df['price'],
    random_state=2025,
    test_size=0.3
)

ridge = Ridge(alpha=1)
ridge.fit(X_train, y_train)
y_preds = ridge.predict(X_test)
```

```python
import pickle

# Сохранение модели в файл
model_filename = "model.pkl"

with open(model_filename, 'wb') as file:
    pickle.dump(ridge, file)
print(f"Модель сохранена в файл: {model_filename}")
```

```python
# Загрузка модели 
with open('model.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
new_data = pd.DataFrame({
    'total_square' : [40.0],
    'rooms' : [2.0]
})
```

```python
predictions = model.predict(new_data)

print('Prediction', predictions)
```
